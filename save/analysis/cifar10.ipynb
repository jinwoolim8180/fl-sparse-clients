```python
from utils import moving_average
import matplotlib.pyplot as plt
import pickle


with open("../results/cifar10/R[300]F[0]LR[0.05]D[noniid]C[100]B[3.0]FR[0.1]_E[3]_LD[0.998].pkl", 'rb') as f:
    fedavg = pickle.load(f)
with open("../results/cifar10/R[300]F[1]LR[0.05]D[noniid]C[100]B[3.0]FR[0.1]_E[3]_LD[0.998].pkl", 'rb') as f:
    feddyn = pickle.load(f)

x = [(i+1) * 5 for i in range(len(fedavg['acc']))][:20]

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_title('noniid (fedavg vs feddyn)')

ax1.set_xlabel("Round")
ax1.set_ylabel('acc', color = 'red')

plt.grid(b=True, which='major', axis='both')
ax1.plot(x, moving_average(fedavg['acc'])[:20], label = 'acc_fedavg', color='orange')
ax1.plot(x, moving_average(feddyn['acc'])[:20], label = 'acc_feddyn', color='red')
ax1.legend(loc='upper right')
ax1.tick_params(axis ='y', labelcolor = 'red') 
ax2 = ax1.twinx()

ax2.set_ylabel('fair', color='blue')
ax2.tick_params(axis ='y', labelcolor = 'blue') 
ax2.plot(x, moving_average([e['diff'] for e in fedavg['fair']])[:20], label = 'fair_fedavg', color='cyan')
ax2.plot(x, moving_average([e['diff'] for e in feddyn['fair']])[:20], label = 'fair_feddyn', color='blue')
ax2.legend(loc='lower right')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pickle


with open("../results/cifar10/R[300]F[0]LR[0.05]D[iid]C[100]B[5.0]FR[0.1]_E[3]_LD[0.996].pkl", 'rb') as f:
    iid = pickle.load(f)
with open("../results/cifar10/R[300]F[0]LR[0.05]D[noniid]C[100]B[0.5]FR[0.1]_E[3]_LD[0.998].pkl", 'rb') as f:
    noniid = pickle.load(f)

x = [(i+1) * 5 for i in range(len(iid['acc']))]

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.set_title('iid vs non iid (fedavg)')

ax1.set_xlabel("Round")
ax1.set_ylabel('acc', color = 'red')

plt.grid(b=True, which='major', axis='both')
ax1.plot(x, iid['acc'], label = 'acc_iid', color='orange')
ax1.plot(x, noniid['acc'], label = 'acc_noniid(1.0)', color='red')
ax1.legend(loc='upper right')
ax1.tick_params(axis ='y', labelcolor = 'red') 
ax2 = ax1.twinx()

ax2.set_ylabel('fair', color='blue')
ax2.tick_params(axis ='y', labelcolor = 'blue') 
ax2.plot(x, [e['diff'] for e in iid['fair']], label = 'fair_iid', color='cyan')
ax2.plot(x, [e['diff'] for e in noniid['fair']], label = 'fair_noniid', color='blue')
ax2.legend(loc='lower right')
plt.show()
```
